```python
import numpy as np 
import pandas as pd 
import os 
for dirname, _, filenames in os.walk('.'): 
    for filename in filenames: 
        print(os.path.join(dirname, filename))
```

```python
items = pd.read_csv('./competitive-data-science-predict-future-sales/items.csv')
item_categories = pd.read_csv('./competitive-data-science-predict-future-sales/item_categories.csv')
test = pd.read_csv('./competitive-data-science-predict-future-sales/test.csv')
shops = pd.read_csv('./competitive-data-science-predict-future-sales/shops.csv')
sales = pd.read_csv('./competitive-data-science-predict-future-sales/sales_train.csv', parse_dates=['date'], 
                   dtype = {'date' : 'str', 'item_cnt_day':'int32'})

```

```python
train_data = sales.join(items, on='item_id', rsuffix ='_').join(shops, on='shop_id', rsuffix='_').join(item_categories, on ='item_category_id', rsuffix='_').drop(['item_id_', 'shop_id_', 'item_category_id_'], axis =1)
```

```python
train_data['item_cnt_day'] = np.abs(train_data['item_cnt_day'])
```

```python
import datetime as dt 

train_data['date'] = pd.to_datetime(train_data['date'], format ='%d.%m.%Y')
train_data['day'] = train_data['date'].dt.day
train_data['month'] = train_data['date'].dt.month
train_data['year'] = train_data['date'].dt.year

display(train_data.head(), train_data)
```

```python
train_data.drop('date', axis = 1, inplace = True)
```
